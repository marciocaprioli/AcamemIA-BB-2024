```python
# Definir o final do semestre
fim_primeiro_semestre = pd.Timestamp('2024-06-30')

# Definir as condições de situação e colaboração
situacoes_diferentes_100 = {328, 310, 311, 222, 481, 482, 483, 484, 485, 461, 315}
colaboracao_keywords = {'DISP', 'ADIDOS', 'CEDIDOS'}

# Função para calcular dias entre duas datas
def calcular_dias(data_inicio, data_fim):
    return max((data_fim - data_inicio).days, 0)

# Função para calcular dias em situação diferente de 100 e dias de colaboração
def calcular_dias_por_situacao(est_fun_situacoes_alvo_1sem):
    est_fun_situacoes_alvo_1sem['dias_diferente_100'] = est_fun_situacoes_alvo_1sem.apply(
        lambda row: calcular_dias(max(row['DT_INC_EST_FUN'], pd.Timestamp('2024-01-01')),
                                  min(row['DT_FIM_EST_FUN'], fim_primeiro_semestre))
        if row['CD_TIP_EST_FUNL'] in situacoes_diferentes_100 else 0, axis=1
    )

    est_fun_situacoes_alvo_1sem['dias_colaboracao'] = est_fun_situacoes_alvo_1sem.apply(
        lambda row: calcular_dias(max(row['DT_INC_EST_FUN'], pd.Timestamp('2024-01-01')),
                                  min(row['DT_FIM_EST_FUN'], fim_primeiro_semestre))
        if any(keyword in row['descricao'] for keyword in colaboracao_keywords) and 'FOLGA' not in row['descricao'] else 0, axis=1
    )

    return est_fun_situacoes_alvo_1sem

est_fun_situacoes_alvo_1sem = calcular_dias_por_situacao(est_fun_situacoes_alvo_1sem)

# Função para determinar se o funcionário é avaliável ou não
def avaliar_funcionario(row):
    dias_diferente_100 = row['dias_diferente_100']
    dias_colaboracao = row['dias_colaboracao']
    total_dias = dias_diferente_100 + dias_colaboracao

    # Verificar as condições
    if dias_diferente_100 > 90:
        return 'N'
    elif dias_colaboracao > 90:
        return 'N'
    elif total_dias > 90:
        return 'N'
    elif row['descricao'] in colaboracao_keywords and row['DT_FIM_EST_FUN'] == fim_primeiro_semestre:
        return 'N'
    elif (pd.Timestamp('2024-06-30') - row['DT_PSSE_BB']).days < 90:
        return 'N'
    else:
        return 'S'

est_fun_situacoes_alvo_1sem['IN_FUN_AVLV_PRMO_SMT'] = est_fun_situacoes_alvo_1sem.apply(avaliar_funcionario, axis=1)
est_fun_situacoes_alvo_1sem.info()
```
