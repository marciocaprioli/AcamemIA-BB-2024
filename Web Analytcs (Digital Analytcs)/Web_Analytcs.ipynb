```python
# 10) Qual o tempo médio em minutos que o cliente permanece em cada visita?
# Regra 1: considere os centésimos de segundo para o cálculo.
# Regra 2: a resposta deverá conter até 2 casas decimais, arredondando sempre para baixo.
# Dica: faça o cálculo da média utilizando todas as casas decimais, arredonde apenas a resposta.

# Calcular a diferença entre o primeiro e o último registro por visita (identificador_visita)
duracoes = usuario_acessos.groupby('identificador_visita')['data'].agg(['min', 'max'])

# Calcular a duração total em minutos para cada visita
duracoes['duracao_minutos'] = (duracoes['max'] - duracoes['min']).dt.total_seconds() / 60

# Calcular a média das durações em minutos
media_duracao = duracoes['duracao_minutos'].mean()

# Arredondar para 2 casas decimais para baixo
media_duracao = media_duracao // 0.01 / 100  # Arredonda para baixo

print(f'A média de tempo de permanência em cada visita é: {media_duracao:.2f} minutos')
```

```python
# 11) Entender a frequência de retorno do cliente ao aplicativo é uma informação
# importante que indica o nível de interesse e pode subsidiar o planejamento de
# estratégias de retenção do usuário.

# Qual o intervalo médio em dias de utilização do app pelo cliente?
# Regra 1: considere a data de início apenas. Exemplo: se a visita começou dia 10/03
# às 11h45 e terminou dia 11/03 às 0h05, considere que a visita ocorreu no dia 10/03.
# Regra 2: caso haja mais de uma visita no mesmo dia, considere apenas uma ocorrência.
# Regra 3: o resultado em dias deverá ser um número inteiro, arredondando para baixo.
# Regra 4: caso o usuário tenha visitado o app apenas um dia, a resposta será null.

# 1. Extrair apenas a data (ignorando a hora)
usuario_acessos['data_unica'] = usuario_acessos['data'].dt.date

# 2. Remover visitas duplicadas no mesmo dia (considerando apenas uma visita por dia)
dias_unicos = usuario_acessos['data_unica'].drop_duplicates()

# 3. Verificar o número de dias únicos em que o cliente acessou o app
num_dias = len(dias_unicos)

# 4. Caso o usuário tenha acessado o app apenas um dia, o resultado será null
if num_dias <= 1:
    resultado = None
else:
    # 5. Calcular o intervalo entre o primeiro e o último dia de visita
    intervalo_dias = (dias_unicos.max() - dias_unicos.min()).days

    # 6. Calcular o intervalo médio e arredondar para baixo
    intervalo_medio = np.floor(intervalo_dias / (num_dias - 1))

    resultado = int(intervalo_medio)  # Garantir que o resultado seja inteiro

# Exibir o resultado
print(f"Intervalo médio de utilização do app: {resultado} dias")
```

```python
# 12) Em qual período o cliente navega com maior frequência?
# Regra 1: Considere os seguintes intervalos para classificação do horário:
# • madrugada: das 00h00 às 05h59m59s
# • manhã: das 06h00 às 11h59 m59s
# • tarde: das 12h00 às 17h59 m59s
# • noite: das 18h00 às 23h59 m59s

# No exemplo acima, o cálculo deverá considerar as ocorrências únicas do período
# por visita

# Regra 2: a resposta deverá obedecer à seguinte classificação: manhã, tarde, noite
# ou madrugada.
# Regra 3: em caso de empate, considere o maior horário, obedecendo à seguinte
# escala: madrugada < manhã < tarde < noite

# 1. Extrair a hora da visita
usuario_acessos['hora'] = usuario_acessos['data'].dt.hour

# 2. Classificar as visitas em períodos do dia
def classificar_periodo(hora):
    if 0 <= hora <= 5:
        return 'madrugada'
    elif 6 <= hora <= 11:
        return 'manhã'
    elif 12 <= hora <= 17:
        return 'tarde'
    elif 18 <= hora <= 23:
        return 'noite'

# 3. Aplicar a classificação para cada visita
usuario_acessos['periodo'] = usuario_acessos['hora'].apply(classificar_periodo)

# 4. Contabilizar as ocorrências únicas por período (considerando identificador_visita como único)
# Não vamos contar visitas duplicadas para o mesmo período
visitas_por_periodo = usuario_acessos.drop_duplicates(subset=['periodo', 'identificador_visita'])['periodo'].value_counts()

# 5. Determinar o período com maior frequência
# Caso haja empate, a prioridade será: madrugada < manhã < tarde < noite
periodos = ['madrugada', 'manhã', 'tarde', 'noite']
periodo_mais_frequente = max(periodos, key=lambda x: (visitas_por_periodo.get(x, 0), -periodos.index(x)))

# Exibir o resultado
print(f"O período de maior frequência de visitas é: {periodo_mais_frequente}")
```

```python
# 13) Há variação nos horários de navegação do cliente nos finais de semana em
# relação ao seu comportamento nos dias úteis?
# Regra1: a resposta deve ser apenas sim ou não.
# Regra2: considere dias não úteis apenas os finais de semana (sábados e domingos).
# Regra3: a forma de cálculo dos horários de maior frequência de navegação seguirá
# as mesmas regras do exercício 12, atentando apenas para a separação entre dias úteis e não úteis.
# Regra4: caso o cliente só tenha acessado o app em uma das situações (dias úteis ou não úteis), a resposta será sim.

# 1. Extrair a hora da visita e o dia da semana
usuario_acessos['hora'] = usuario_acessos['data'].dt.hour
usuario_acessos['dia_semana'] = usuario_acessos['data'].dt.weekday  # 0 = segunda, 6 = domingo

# 2. Classificar as visitas em períodos do dia
def classificar_periodo(hora):
    if 0 <= hora <= 5:
        return 'madrugada'
    elif 6 <= hora <= 11:
        return 'manhã'
    elif 12 <= hora <= 17:
        return 'tarde'
    elif 18 <= hora <= 23:
        return 'noite'

usuario_acessos['periodo'] = usuario_acessos['hora'].apply(classificar_periodo)

# 3. Separar as visitas em dias úteis e finais de semana
dias_uteis = usuario_acessos[usuario_acessos['dia_semana'] < 5]  # Dias úteis: segunda a sexta
finais_de_semana = usuario_acessos[usuario_acessos['dia_semana'] >= 5]  # Sábado e domingo

# 4. Contabilizar as visitas únicas por período para dias úteis e finais de semana
# Para cada conjunto, contamos as visitas únicas por período
visitas_uteis = dias_uteis.drop_duplicates(subset=['periodo', 'identificador_visita'])['periodo'].value_counts()
visitas_fds = finais_de_semana.drop_duplicates(subset=['periodo', 'identificador_visita'])['periodo'].value_counts()

# 5. Determinar o período de maior frequência para dias úteis e finais de semana
# Caso o usuário não tenha acessado algum período, o valor retornado será zero (não foi acessado naquele período)
periodos = ['madrugada', 'manhã', 'tarde', 'noite']
periodo_uteis = max(periodos, key=lambda x: (visitas_uteis.get(x, 0), -periodos.index(x)))
periodo_fds = max(periodos, key=lambda x: (visitas_fds.get(x, 0), -periodos.index(x)))

# 6. Comparar os períodos de maior frequência
if periodo_uteis != periodo_fds:
    resposta = "sim"
else:
    # 7. Caso o cliente só tenha acessado o app em um tipo de dia (dias úteis ou finais de semana)
    if len(visitas_uteis) == 0 or len(visitas_fds) == 0:
        resposta = "sim"
    else:
        resposta = "não"

# Exibir a resposta
print(f"Há variação nos horários de navegação? {resposta}")
```

```python
# 22) Qual a etapa do funil de conversão do produto onde se observa maior taxa de
# abandono?
# Regra 1: considere cada página pré-conversão uma etapa do funil.
# Regra 2: para calcular, identifique qual foi a última página de pré-conversão
# acessada nas visitas em que não houve conversão.

# Lista de páginas de pré-conversão e pós-conversão
paginas_pre_conversao = ['confirmacao_boleto', 'pagamento']
pagina_pos_conversao = 'sucesso_boleto'

# Passo 1: Filtrar as visitas que acessaram as páginas de pré-conversão
visitas_pre_conversao = filtro_produto[filtro_produto['pagina_acessada'].isin(paginas_pre_conversao)]

# Passo 2: Identificar as visitas sem conversão (não acessaram a página de pós-conversão)
visitas_sem_conversao = visitas_pre_conversao[~visitas_pre_conversao['identificador_visita'].isin(
    filtro_produto[filtro_produto['pagina_acessada'] == pagina_pos_conversao]['identificador_visita']
)]

# Passo 3: Identificar a última página de pré-conversão acessada nas visitas sem conversão
# Para isso, ordenamos as visitas por 'identificador_visita' e 'data', e selecionamos a última página acessada
visitas_sem_conversao = visitas_sem_conversao.sort_values(by=['identificador_visita', 'data'])

# Agora, para cada visita, pegamos a última página de pré-conversão acessada
ultima_pagina_pre_conversao = visitas_sem_conversao.groupby('identificador_visita').last().reset_index()

# Passo 4: Calcular a taxa de abandono
# Contabilizar o total de visitas que acessaram as páginas de pré-conversão
total_visitas_pre_conversao = visitas_pre_conversao.drop_duplicates(subset=['identificador_visita'])

# A taxa de abandono será o número de visitas sem conversão dividido pelo total de visitas com pré-conversão
taxa_abandono = len(visitas_sem_conversao) / len(total_visitas_pre_conversao)

# Passo 5: Identificar a etapa do funil com maior taxa de abandono
# Contabilizamos as visitas sem conversão para cada página de pré-conversão
abandono_por_etapa = ultima_pagina_pre_conversao['pagina_acessada'].value_counts(normalize=True)

# A página com maior taxa de abandono será a etapa mais problemática do funil
maior_abandono_etapa = abandono_por_etapa.idxmax()
maior_abandono_taxa = abandono_por_etapa.max()

# Exibindo os resultados
print(f"A taxa de abandono da conversão do produto é: {round(taxa_abandono, 2)}")
print(f"A etapa do funil com maior taxa de abandono é: {maior_abandono_etapa} com taxa de abandono de: {round(maior_abandono_taxa, 2)}")
```

```python
# 28) Qual a taxa de variação semanal média do número de visitas únicas em setembro/2024?
# Regra 1: considere as 4 semanas cheias do mês:
# • semana 1: 01 a 07/09
# • semana 2: 08 a 14/09
# • semana 3: 15 a 21/09
# • semana 4: 22 a 28/09

# Regra 2: a reposta será a média da taxa de variação entre cada período.
# Regra 3: o valor deverá ser representado de forma decimal, com até 2 casas
# decimais, arredondando para o valor mais próximo.
# Dica 1: a variação entre dois períodos é calculada da seguinte forma: o total de
# visitas do período inicial menos o total de visitas do período final, dividido pelo
# total de visitas do período inicial.
# Dica 2: arredonde apenas a resposta final

# Filtrando os acessos de setembro de 2024
acessos_sept = acessos[(acessos['data'].dt.month == 9) & (acessos['data'].dt.year == 2024)]

# Definir as datas que delimitam as semanas
semanas = {
    'semana_1': ('2024-09-01', '2024-09-07'),
    'semana_2': ('2024-09-08', '2024-09-14'),
    'semana_3': ('2024-09-15', '2024-09-21'),
    'semana_4': ('2024-09-22', '2024-09-28'),
}

# Função para contar visitas únicas por semana
def contar_visitas_unicas(start_date, end_date):
    visitas_semana = acessos_sept[(acessos_sept['data'] >= start_date) & (acessos_sept['data'] <= end_date)]
    return visitas_semana['identificador_visita'].nunique()

# Contar visitas únicas por semana
visitas_semanas = {}
for semana, (start, end) in semanas.items():
    visitas_semanas[semana] = contar_visitas_unicas(pd.to_datetime(start), pd.to_datetime(end))

# Calcular as taxas de variação entre as semanas
taxas_variacao = []
for i in range(1, len(visitas_semanas)):
    semana_inicial = list(visitas_semanas.keys())[i-1]
    semana_final = list(visitas_semanas.keys())[i]

    visitas_inicial = visitas_semanas[semana_inicial]
    visitas_final = visitas_semanas[semana_final]

    taxa_variacao = (visitas_final - visitas_inicial) / visitas_inicial
    taxas_variacao.append(taxa_variacao)

# Calcular a média das taxas de variação
media_taxa_variacao = sum(taxas_variacao) / len(taxas_variacao)

# Arredondar para 2 casas decimais
media_taxa_variacao = round(media_taxa_variacao, 2)

print(f'A taxa de variação semanal média é: {media_taxa_variacao}')
```
